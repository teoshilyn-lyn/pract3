## Распознование символов

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.layers import Dense, Flatten
```

### Создаем данные, нужные для обучения и проверки
### Рисуем буквы латинского алфавита на холсте размеров 3x3 пиксела

```python
X = np.array([
    
    [[1, 0, 1],
     [0, 1, 0],   # Буква X
     [1, 0, 1]],
    
    [[1, 0, 1],
     [0, 1, 0],   # Буква Y
     [0, 1, 0]],
    
    [[0, 1, 0],
     [0, 1, 0],   # Буква I
     [0, 1, 0]],
    
    [[1, 1, 1],
     [1, 0, 0],   # Буква C
     [1, 1, 1]]
])
```

### Если мы передадим на выход сети просто одно число, то сети будет очень трудно правильно распознать границы допустимых значений, да решения многоклассовой классификации есть другой способ
### Вместо одного выходного нейрона мы используем 4 и применим к ним функцию softmax, которая переведет все их значения в такие, что сумма этих 4 значений будет равняться 1
### После чего выберем самое большое значение из 4, оно и будет выбором сети

```python
num_word = {
    0: 'X',
    1: 'Y',
    2: 'I',
    3: 'C'
}

word_num = {
    'X': np.array([1, 0, 0, 0]),
    'Y': np.array([0, 1, 0, 0]),
    'I': np.array([0, 0, 1, 0]),
    'C': np.array([0, 0, 0, 1])
}

y = np.array([
    word_num['X'],
    word_num['Y'],
    word_num['I'],
    word_num['C']
])
```

### Создаем модель нейронной сети с помощью класса keras.Sequential, что переводится как последовательность, так как слои в этой нейронной сети идут ровно друг за другом (бывают и другие архитектуры нейронных сетей)
### Потом мы определяем функцию градиентного спуска (optimizer='rmsprop') функцию потерь, по которой модель будет понимать на сколько она ошибается, и метрику качества, которая нужно только нам. Все это делается в методе model.compile
### В этот раз первым слоем мы прописали Flatten, все, что он делает, это разворачивает нашу матрицу значений в вектор и передает его дальше, самый простой слой, который только можно придумать
### Выходной частью нашей сети будет четыре нейрона с функцией активации softmax, как уже говорилось выше, сумма эти 4 значений будет равняться 1 и нам нужно просто самое большое из них

```python
model = keras.Sequential()
model.add( Flatten(input_shape=(3, 3)) )
model.add( Dense(16, activation='relu') )
model.add( Dense(16, activation='relu') )
model.add( Dense(4, activation='softmax') )

model.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics='accuracy'
)

model.fit(
    x=X,
    y=y,
    epochs=5
)
```

### Предскажем букву I
### np.expand_dims создает еще одно измерение, когда мы обучали сеть то передовали в нее вектор матриц, так что теперь ничего кроме вектора матриц наша сетка не примет, по этому когда мы взяли изображение лишь одной цифры ее нужно обернуть в вектор

```python
predict = model.predict( np.expand_dims(X[2, :, :], axis=0) )

predict
```

### Сеть вернула нам значения каждого из выходных нейронов, теперь найдем самое большое из них

```python
result = np.argmax(predict)

num_word[result]
```

### Сети удалось правильно распознать цифру, ура)
