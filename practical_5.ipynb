## Создание и обучение простейшей нейронной сети для классификации исключающего или

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

### Создаем данные, нужные нам для обучения и проверки
### Проверять сеть буде на тех же данных, ибо варианты сочетаний цифр 0 и 1 уже исчерпаны

```python
X_train = np.array([[0, 0],
                    [0, 1],
                    [1, 0],
                    [1, 1]])

y_train = np.array([1, 0, 0, 1])
```

### Создаем модель нейронной сети с помощью класса keras.Sequential, что переводится как последовательность, так как слои в этой нейронной сети идут ровно друг за другом (бывают и другие архитектуры нейронных сетей)
### Потом мы определяем функцию градиентного спуска (optimizer='adam') функцию потерь, по которой модель будет понимать на сколько она ошибается, и метрику качества, которая нужно только нам. Все это делается в методе model.compile
### И, наконец, просто обучам сеть на наших данных

```python
model = keras.Sequential()
model.add( Dense(2, activation='relu', input_shape=(2,)) )
model.add( Dense(2, activation='relu') )
model.add( Dense(1, activation='sigmoid') )

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics='accuracy'
)

model.fit(
    x=X_train,
    y=y_train
)
```

### С помощью метода model.evaluate мы предсказываем значения для новых данных и сразу же проверяем их по метрике, переданной в методе model.compile

```python
model.evaluate(X_train, y_train)
```

### Это можно объяснить тем, что нейросеть пытается построить сложную внутреннюю структуру переданных в нее данных и, пока она "умничает", упускает банальные и очень простые паттерны классификации
